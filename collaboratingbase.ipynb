```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv('Books.csv')
rating = pd.read_csv('Ratings.csv')
user = pd.read_csv('Users.csv')
```

```python
book.head(2)
```

```python
rating_w_book = rating.merge(book , on='ISBN') 
rating_w_book
```

#now only those user stay which has min 200 votes

```python
rating_w_book.groupby('User-ID').count()['Book-Rating']  #92106 user who rate the books
```

```python
x =  rating_w_book.groupby('User-ID').count()['Book-Rating'] > 200 # 'User-ID' count with Book-Rating
#x
study_user = x[x].index #user who send the reveiw more than 200 books
study_user
```

```python
filter_rating = rating_w_book[rating_w_book['User-ID'].isin(study_user)]
filter_rating
```

```python
y = filter_rating.groupby('Book-Title').count()['Book-Author'] > 50

famous_book = y [y].index #more than 50 ratings books
famous_book  
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_book)]
final_rating  
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index='Book-Title' , columns='User-ID' , values='Book-Rating')
pt
#here in pivot table books is more than 50 ratings
# and user has more than 250 review on books
```

```python
pt.fillna(0 , inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similar_score = cosine_similarity(pt)#vector 2d numpy array 
similar_score 
```

```python
cosine_similarity(pt).shape #we find 679 books  Distance with the 679 books
```

```python
similar_score[0] #this 1st Array said that the first book -> the Score of 1st books(1984) to others book
# and that the reason of same shape(size) of similar_score(679 , 679)
```

```python
def recommend(book_name):
    #index fetch 
    index = np.where(pt.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similar_score[index])) , key=lambda z:z[1] , reverse=True) [1:6]
    data = []
    for i in similar_item:
       # print(pt.index[i[0]])
       item = []
       temp_df = book[book['Book-Title'] == pt.index[i[0]]]
       item.extend(temp_df.drop_duplicates('Book-Title')['Book-Title'].values)
       item.extend(temp_df.drop_duplicates('Book-Title')['Book-Author'].values)
       item.extend(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values)
       item.extend(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values)
       
       data.append(item)

    return data
#pt.index --> all books name and i[0] perticular 
```

enumerate -> find the index of item
key = lambda z:z[1] --> used to perform sorting on distance not index below
[1:6] --> used to only top 5
